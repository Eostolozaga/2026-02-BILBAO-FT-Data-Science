## Juego del Ahorcado
El funcionamiento es que vaya mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _```. 

Si falla, se mostrará el número de errores que lleva y, si lleva 6 errores, entonces se mostrará "Fin del juego".
Encapsúlalo en una función.

````python
def ahorcado_start():
    print(" Bienvenido al juego del Ahorcado.Tendrás 6 vidas para intentar adivinar la palabra.¿Aceptas el reto?")
    
    while True:
        entrada = input("Elige una opción (0 = Salir, 1 = Jugar): ").strip()
        
        # Si pulsa Enter sin escribir nada
        if not entrada:
            print(" No has introducido nada. Intenta de nuevo.")
            continue
        
        # Comprobar si el texto es numérico
        if not entrada.isdigit():
            print(" Debes introducir un número entero (0 o 1).")
            continue
        
        comienzo = int(entrada) 
        
        if comienzo == 0:
            print("Gracias por jugar. ¡Hasta pronto!")
            return
        elif comienzo == 1:
            iniciar_juego()
            return
        else:
            print("Opción inválida. Introduce 0 o 1.")
           
    
````

````python
import random # Para que pueda elegir el programa de forma aleatoria que palabra poner y no sea tan limitado
````

````python
class Tablero:                                              # Me gustaria importar un diccionario o algo similar en vez de una lista limitada, pero no he sabido hacerlo
    palabras = ["Glacial", "Trabajador", "Dirigible", "Altercado", "Agresión", "Infestar", "Autopsia", "Adicto", "Rico", "Divina", "Fragmento", "Prometido", "Devoto", "Sirviente", "Tablero", "Viga", "Reflexivo", "Membrillo", "Tendencia"]   
    
    def __init__(self):
        self.palabra_secreta = None
        self.tablero = None
        self.letras_usadas = set()
        self.palabra_secreta = random.choice(self.palabras).lower() # Escoge al azar una palabra de la lista
        self.longitud = len(self.palabra_secreta) 
        self.tablero = ["_" for _ in range(self.longitud)]
        
    def mostrar_tablero(self):
        print(" ".join(self.tablero))
        
    def procesar_letra(self, letra):
        acierto = False
        for i, letra_secreta in enumerate(self.palabra_secreta):
            if letra_secreta == letra:
                self.tablero[i] = letra
                acierto = True
        return acierto

    def esta_completa(self):
        return "_" not in self.tablero
    
````
