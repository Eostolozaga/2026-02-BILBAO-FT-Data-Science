## Ejercicio 1
Declara una variable numérica que será una hora cualquiera del día. Implementa la siguiente casuística:

* Si es entre las 0 y las 8, print "Durmiendo"
* Si es entre las 9 y las 18, print "Trabajando"
* Si es entre las 19 y las 21, print "Clase"
* Si es entre las 22 y las 24, print "Descanso"
* En cualquier otro caso, print "Transporte o error"

```python
hora_dia = -12

if hora_dia <= 0 or hora_dia >24:
    print(f"{hora_dia}, Esas horas están fuera del rango")            # Primero nos adelantamos a posibles errores y acotamos las horas inexistentes
elif hora_dia >= 0 and hora_dia <= 8:                                 # Acotamos el resto de horas en la horquilla indicada
    print("Durmiendo")
elif hora_dia <= 9 and hora_dia <= 18:
    print("Trabajando")
elif hora_dia <= 19 and hora_dia <= 21:
    print("Clase")
elif hora_dia <= 22 and hora_dia <= 24:
    print("Descanso")                                                # Creo que en este caso no es necesario un else puesto que ya hemos cubierto las variables en el primer if
    
```

## Ejercicio 2
En este ejercicio vamos a implementar un calculador de precios de casas muy sencillo. Tenemos las siguientes variables:

> superficie
>
> distrito

Implementa la siguiente casuística:

1. Si el distrito es "Moncloa" o "Centro", y además la superficie es superior a 100 metros cuadrados, el precio de la casa es de 1000
2. Si el distrito es "Salamanca", y además la superficie de la casa es al menos de 150 metros, el precio de la casa es de 1500
3. Si el distrito no es "Retiro" y la superficie está entre 60 y 80 metros, el precio es de 600
4. En cualquier otro caso, el precio será de 0

```python
distrito = input("Introduce uno de los distritos")
superficie = input("Introduce una superficie")

if distrito.lower == "moncloa" or "centro" and float(superficie) >= 100:                                # Usamos un lower para evitar errores de mayusculas o diferentes formatos
    print(f"El precio de una casa en el distrito de {distrito}, con una superficie de {superficie}, es de 1000")
elif distrito.lower == "salamanca" and float(superficie) >= 150:
    print(f"El precio de una casa en el distrito de {distrito}, con una superficie de {superficie}, es de 1500")
elif distrito.lower == "retiro" and float(distrito) >= 60 and float(distrito)<= 80:
    print(f"El precio de una casa en el distrito de {distrito}, con una superficie de {superficie}, es de 600")
else:
    print("El precio es de 0")
```

## Ejercicio 3
En este ejercicio vamos a realizar un programa muy parecido. Para este caso queremos que se cumplan las siguientes condiciones:

1. Primero se compruebe si el distrito es "Retiro". Si es asá, que imprima "Distrito Retiro", y si no, "Otro distrito"
2. Si el distrito es "Retiro", comprueba si la superficie es mayor de 100 metros cuadrados. En tal caso, que imprima un precio de 1000, y si no, de 500.

Prueba a usar `ifs` anidados

```python
distrito_introducido = input("Introduce un distrito")
superficie = input("Dame una superficie")

if distrito_introducido.lower() == "retiro":  # definimos "retiro" como condición y aplicamos el lower para evitar tema de mayusculas
    if float(superficie) >= 100:    # Float para evitar problemas
        print(f"El precio para una casa en {distrito_introducido}, con superficie de {superficie}, es de 1000")
    elif float(superficie) >= 0 and float(superficie) <= 100: # Ponemos una horquilla para datos reales, tiene que ser superior a 0
        print(f"El precio para una casa en {distrito_introducido}, con superficie de {superficie}, es de 500")
    else:
        print("El valor introducido no es correcto") # El resto de opciones
if distrito_introducido.lower() != "retiro":        # Si es distinto a retiro directamente devolvemos el output de que no es el distrito que estamos pidiendo
    print(f"{distrito_introducido} es otro distrito")
    
    
```

## Ejercicio 6
Dada la siguiente lista:

> list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

Implementa un programa que los recorra e imprima por pantalla todos los divisibles por 5. Si nos encontramos con alguno que sea mayor que 150, detener el bucle.

```python
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]
for numero in (list1):
    if numero > 150:                # Ponemos primero el condicionante más fuerte para que pare antes del print y no enseñe 180
        break
    elif numero % 5 == 0:
        print(numero)
```

## Ejercicio 7
Escribe un programa en Python que imprima por pantalla todos los números que cumplan que sean divisibles por 5 y divisibles por 7, dentro del rango de valores (150, 350).

## Ejercicio 8

5 4 3 2 1 

4 3 2 1 

3 2 1 

2 1

NOTA: NO hay lineas en blanco entre una línea y otra.

```python
dibujo = [list(range(5,0,-1)),list(range(4,0,-1)),list(range(3,0,-1)),list(range(2,0,-1)),list(range(1,0,-1))]  # Las 5 secuencias de la foto
for linea in dibujo:                                                                                            # Bucle que itera sobre cada linea
    for num in linea:                                                                                           # Bucle que itera sobre cada numero
        print(num, end=' ')                                                                                     # Sirve para controlar que se imprime al final
    print()                                                                                                     # Crea el salto de linea


```

## Ejercicio 9
En este ejercicio vamos a crear un pequeño juego. Se trata de intentar adivinar un numero del 1 al 5. Tenemos dos intentos para acertar. Pasos a seguir:

1. Ya viene implementado con la librería `random` cómo obtener un número aleatorio del 1 al 5
2. Tendrás que declarar en una variable el numero de vidas, y mediante un bucle while, comprobar que todavia quedan vidas.
3. Dentro del bucle, obtener el valor del usuario y comprobar si es ese el numero a adivinar. Si no, actualizar las vidas.
4. Si acertamos, salimos del bucle e imprimimos por pantalla "You win". Y si perdemos también salimos del bucle, pero en este caso imprimimos por pantalla "You lose".

TIP: te puede resultar útil usar la sentencia `else` cuando acabe el bucle `while`. Lo que haya dentro de ese `else` se ejecutará una vez acabe la ejecución del `while`. Lo podrás usar para cuando pierdas.

```python
import random
```

```python
num_adivinar = random.randint(1,5)                                               # Generador random de numeros entre 1 y 5
num_vidas = 2
while num_vidas > 0:                                                             # Primera condicion que queden vidas restantes
    juego = int(input("Te desafio! Intenta adivinar mi número del 1 al 5"))      # Input dentro del bucle para que pida por si falla
    if juego == num_adivinar:
        print(f"{num_adivinar} era mi numero. You win!")
        break                                                                     # En caso de acierto salimos del bucle
    else:                                                                         # Si falla tenemos el comando de quitar una vida. 
        num_vidas = (num_vidas) - 1                                               # Se vuelve a ejecutar hasta que en el while la vida sea de 0
        print(f"{juego} no es mi número. Has perdido una vida. Numero de vidas {num_vidas}")
print(f"{num_vidas} vidas restantes. You lose!")                                    # Fuera del bucle para que salte cuando el while alcance 0.
```
