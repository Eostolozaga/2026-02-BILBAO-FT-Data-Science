## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
print("Bienvenido a tu calculadora, estas son las opciones disponibles")   # Asigno a cada numero una funcion y se la muestro al usuario.
print("1.Sumar")                                                           # Seguramente se puedan definir tambien, pero de forma más compleja
print("2.Restar")
print("3.Multiplicar")
print("4.Dividir")
print("5.Potencia")
print("6.Salir")
while True:                                                                # Bucle infinito
    operacion = float(input("elige una operación del 1 al 6"))
    if operacion == 6:
        print("Cerrando calculadora")
        break                                                               # Como otras veces ponemos la opción más restrictiva la primera
    num1= float(input("Elige el primer numero"))                            # Pedimos los dos inputs al usuario una vez comprobado que no desea salir
    num2 = float(input("Elige el segundo numero"))
    if operacion == 1:
        print(f"El resultado de sumar {num1} y {num2} es de {num1 + num2}")
    elif operacion == 2:
        print(f"El resultado de restar {num1} y {num2} es de {num1 - num2}")
    elif operacion == 3:
        print(f"El resultado de multiplicar {num1} y {num2} es de {num1 * num2}")
    elif operacion == 4:
        if num1 != 0 and num2 != 0:
            print(f"El resultado de dividir {num1} y {num2} es de {num1 / num2}")
        else:
            print("Error, no se puede dividir por 0")
    elif operacion == 5:
        print(f"El resultado de la potencia de {num1} y {num2} es de {math.pow(num1,num2)}")        # Usamos la libreria math para poder hacer la potencia
        
# He intentado que al menos las divisiones de 0 se eviten pero no se almacenar y devolver las operaciones anteriores
    
```
