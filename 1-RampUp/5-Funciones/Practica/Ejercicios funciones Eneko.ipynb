## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def dias_semana(numero):
    if numero == 1:                                     # Defino todos los días de la semana asignando un número
        return ("Lunes")
    elif numero == 2:
        return("Martes")
    elif numero == 3:
        return("Miercoles")
    elif numero == 4:
        return("Jueves")
    elif numero == 5:
        return ("Viernes")
    elif numero == 6:
        return ("Sabado")
    elif numero == 7:
        return ("Domingo")
    else:
        return ("No es un numero valido")
        
````

````python
dias_semana(7)
````

## Ejercicio 2
En el ejercicio 8 de flujos de control, creábamos una pirámide invertida de números desde el 5. Crea una función que replique el comportamiento de la pirámide, y utiliza un único parámetro de entrada de la función para determinar el número de filas de la pirámide.

````python
def piramide(filas):
    for i in range(filas,0,-1):  # Rango de el input de filas, stop en el 0 y vaya quitando una unidad (5,4,3,2,1)
        for j in range(i,0,-1):  # Rango de j hasta i decreciendo (Por ej., cuando i es 5, j sera 5,4,3,2,1)
            print(j, end = "")     # El end = " " como vimos en una clase anterior sirve para evitar el salto de linea generico del print
        print()
````

````python
piramide(5)
````

## Ejercicio 3
Escribe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero sea  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def numeros_iguales(numero1,numero2):           
    if numero1 == numero2:                                                      
        return ("Los numeros son iguales")
    elif numero1 != numero2:                                            # Ifs anidados para cuando se cumpla la condición de que sean diferentes valore si es mayor o no
        if numero1 > numero2:
            return (f"Los numeros {numero1} y {numero2} son distintos y {numero1} es mayor")
        else:
            return (f"Los numeros {numero1} y {numero2} son distintos")
````

````python
numeros_iguales(16,7)
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea una letra que contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayúscula, como minúscula

````python
def contador_letras(texto,letra):
    if letra in texto:
        return (texto.lower ().count(letra), f"veces encontrada la letra {letra}")
    else:
        return ("No se ha encontrado la letra en el texto")
````

````python
contador_letras("Hola que tal","a")
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string, siendo la clave la letra y el valor el conteo.

````python
def contar_letras(texto):
    conteo = {}                                                     # Creamos un diccionario vacío
    for caracter in texto:
        if caracter in conteo:                                      # Si el caracter está en el conteo sumamos 1
            conteo[caracter] = conteo[caracter] + 1
        else:
            conteo[caracter] = 1
    return conteo    
               
````

````python
resultado = contar_letras("Cuenta las letras de este string")
print(resultado)
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

Tendrá como salida la lista

````python
def lista_cambiante(lista1,comando,elemento=None):
    if comando == "add":                                # Establecemos los dos comandos que pueden darse, "add" o "remove"
        if elemento is not None:
            lista1.append(elemento)                     # Si el elemento no es None, se añade mediante append
        else:
            print("Error. Especifica un elemento de la lista")
    elif comando == "remove":
        if elemento is not None and elemento in lista1: # Lo inverso a lo anterior, si no es None, se puede eliminar
            lista1.remove(elemento)
        else:
            print("Error. Especifica un elemento de la lista")
    else:
        print("Error. El comando debe ser add o remove")
    
    return lista1
````

````python
lista_test = ["test1","test2"]
print(lista_test)
lista_test = lista_cambiante(lista_test,"add","bananna")
print(lista_test)
````

## Ejercicio 8
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

````python
def fibonacci(n):                                                   # Fn = Fn-1+ Fn-2
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
````

````python
print(fibonacci(11))
````

Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el área de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math                                                                             # Cuadrado lado * lado      

def area_cuadrado(lado):                                                                # Triangulo base * altura /2
    return lado **2

def area_triangulo(base, altura):                                                       # Circulo radio elevado a 2 * pi
    return (base*altura)/2

def area_circulo(radio):
    return math.pi * radio ** 2    
````

````python
ejercicio1 = (area_circulo(10) * 2) + area_triangulo(3,7)
print(ejercicio1)

ejercicio2 = area_cuadrado(10) + area_circulo(4) + (area_circulo(6)*2) + (area_triangulo(2,4)*5)
print(ejercicio2)
````
