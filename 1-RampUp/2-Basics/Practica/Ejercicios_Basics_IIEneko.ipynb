## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input_1 = input("Mete un input")
input_2 = input("Ahora mete otro")
input_3 = input("Dame un tercer elemento")

if input_1 == input_2 and input_1 == input_3 and input_2 == input_3:                        # Condicional si los 3 inputs son iguales
    print(f"{input_1}, {input_2}, {input_3},Es el mismo input, plagiador!")
elif input_1 == input_2 or input_1 == input_3 and input_1 == input_2 or input_2 == input_3:     # Condición de si 2 de 3 inputs son iguales
    print(f"{input_1}, {input_2} , {input_3} , Al menos hay un elemento distinto, enhorabuena")
else:                                                                                           # Ultima condicion posible
    print(f"{input_1}, {input_2} y {input_3} son diferentes")                       
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
input_1 = float(input("Dame un input"))
input_2 = float(input("Dame un segundo input"))
Test = 10 #Defino el "10" para darle versatilidad por si quisieramos cambiar de número
suma_inputs = input_1 + input_2 #Suma inputs
if suma_inputs > Test:                                                                     # Condicional de si la suma > que el número propuesto
    print(f"La suma de {input_1} y {input_2} es {suma_inputs}, es mayor que {Test}")        
else:                                                                                      # El output si no se cumplen condiciones
    print(f"La suma de {input_1} y {input_2} es{suma_inputs} es inferior a {Test}")
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "Lo primero de todo, cómo están los máquinas"

1. Pasarlo todo a mayúsculas
2. Pasarlo todo a minúsculas
3. Pasar a mayúsculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
text = "Lo primero de todo, cómo están los máquinas"
text_mayus = text.upper             # Upper para mayusculas
print(text_mayus)

text_minus = text.lower             # Lower para mayusculas
print(text_minus)

text_iniciales = text.title()          # Title para que sea solo la primera letra de cada palabra en mayusculas
print(text_iniciales)

text_espacios = text.split()           #Split para dividir las palabras por espacio
print(text_espacios)

text_sustituido = text.replace(",",".") # Replace de "," y "."
print(text_sustituido)

text_eliminar = text.replace("a","")    # Cambiamos la "a" por un espacio lo cual equivale a eliminarlo. Habría que añadir alguna función para que elimine las que tienen tilde.
print(text_eliminar)
````
