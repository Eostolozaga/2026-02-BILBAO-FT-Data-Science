2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
# 1. Valores del eje X entre 0 y 5 (con precisión de 100 puntos)
x = np.linspace(0, 5, 100)

# 2. Definir las funciones
y1 = 3 * x          # Pendiente 3, pasa por (0,0)
y2 = -3 * x - 5     # Pendiente -3, pasa por (0,-5)

# 3. Crear la figura y las dos cajas (subplots)
plt.figure(figsize=(10, 4))

# Gráfica izquierda
plt.subplot(1, 2, 1) # (filas, columnas, índice)
plt.plot(x, y1, color='blue')
plt.title("Pendiente positiva (3x)")
plt.grid(True)

# Gráfica derecha
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red')
plt.title("Pendiente negativa (-3x - 5)")
plt.grid(True)

plt.tight_layout() # Ajusta el espacio entre gráficas
plt.show()
```

3. Fija el eje X entre 0 y 5 y el eje Y entre -15 y 15

```python
# 1. Datos con buena precisión
x = np.linspace(0, 5, 100)
y1 = 3 * x
y2 = -3 * x - 5

plt.figure(figsize=(10, 4))

# Gráfica Izquierda
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue', label='y = 3x')
plt.title("Pendiente positiva")
plt.xlim(0, 5)    # Fija eje X
plt.ylim(-15, 15) # Fija eje Y
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Gráfica Derecha
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red', label='y = -3x - 5')
plt.title("Pendiente negativa")
plt.xlim(0, 5)    # Fija eje X
plt.ylim(-15, 15) # Fija eje Y
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```

4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
x = np.linspace(0, 5, 100)
y1 = 3 * x
y2 = -3 * x - 5

plt.figure(figsize=(10, 5))

# Gráfica Izquierda
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("recta sube")
plt.xlabel("eje X")
plt.ylabel("eje Y") # Esta es la única etiqueta Y que se mostrará
plt.xlim(0, 5)
plt.ylim(-15, 15)
plt.grid(True, alpha=0.3)

# Gráfica Derecha
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red')
plt.title("recta baja")
plt.xlabel("eje X")
# No añadimos plt.ylabel() aquí para que quede limpio
plt.xlim(0, 5)
plt.ylim(-15, 15)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
# 1. Definir datos con numpy
x = np.linspace(0, 5, 100)
y_sube = 3 * x          # Pendiente 3, pasa por (0,0)
y_baja = -3 * x - 5     # Pendiente -3, pasa por (0,-5)

# 2. Crear FIGURA y EJES por separado (1 fila, 2 columnas)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# --- Gráfica Izquierda (Eje 0) ---
axs[0].plot(x, y_sube, color='blue')
axs[0].set_title('recta sube')
axs[0].set_xlabel('eje X')
axs[0].set_ylabel('eje Y') # Solo una etiqueta de eje Y
axs[0].set_xlim(0, 5)
axs[0].set_ylim(-15, 15)
axs[0].grid(True, alpha=0.3)

# --- Gráfica Derecha (Eje 1) ---
axs[1].plot(x, y_baja, color='red')
axs[1].set_title('recta baja')
axs[1].set_xlabel('eje X')
# No añadimos set_ylabel para cumplir con tu petición
axs[1].set_xlim(0, 5)
axs[1].set_ylim(-15, 15)
axs[1].grid(True, alpha=0.3)

# Ajustar diseño
plt.tight_layout()
plt.show()
```

7. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es más sencillo dejarlos dentro de la figura.

```python
personas = ["Ana", "Luis", "Pedro", "Sofía", "Carmen"]
alturas = [160, 180, 175, 190, 170]

serie = pd.Series(alturas, index=personas).sort_values(ascending=False)

plt.bar(serie.index, serie.values, color="green")
plt.xlabel("Personas")
plt.ylabel("Alturas")
plt.title("Comparación alturas grupo")
plt.ylim(140,200);
```
