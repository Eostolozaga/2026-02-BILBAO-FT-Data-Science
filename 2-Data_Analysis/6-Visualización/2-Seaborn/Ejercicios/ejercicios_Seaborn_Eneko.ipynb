# 19. Veamos la función melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
### Dado un DataFrame, melt recibe varios parámetros:
### * qué columnas del DataFrame se quedan igual
### * qué nombres de columnas del dataframe pasan a ser una columna nueva
### * cómo queremos llamar esas nuevas columnas

### Primero recuerda el Dataframe df_grass_fire_water con las primeras 5 filas

```python
pokemon.head()
```

```python
pokemon_melted = pd.melt(pokemon_filtrado, 
                         id_vars=["Name", "Type 1"], # Estas son las que se mantienen
                         var_name="Stat",
                         value_name="Value") 

pokemon_melted.head()
```

# 20. Queremos dejar igual las variables 'Name' y 'Type 1'
### Queremos añadir una columa nueva cuyos valores sean las columnas que queremos apilar y que se llame 'New Column'
### Escribimos df_gfw.melt(id_vars=['Name', 'Type 1'], value_vars=[...], var_name='New Column')

```python
df_melted = pokemon_filtrado.melt(id_vars=['Name', 'Type 1'], 
                        value_vars=['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'], 
                        var_name='New Column', 
                        value_name='Value')

# Comprobamos el resultado
df_melted.head()
```

# 21. Pintamos el swarmplot con el melted y entendemos la información

```python
# Configuramos el estilo
sns.set_style('whitegrid')

# Dibujamos el swarmplot
plt.figure(figsize=(10,6))
sns.swarmplot(x='New Column', y='Value', data=df_melted, 
              hue='Type 1', palette=pokemon_colores, dodge=True)

plt.title('Estadísticas de Pokémon por Tipo (Grass, Fire, Water)')
plt.show()
```

# 23. Mapas de calor
### Calcula las correlaciones del primer DataFrame original df
### Pista: las correlaciones se calculan con el método corr()
### Recuerda: las correlaciones miden entre dos variables, si varío una, cuánto me varía la otra (cómo están de relacionadas)

```python
# Calculamos la matriz de correlación solo para los números
corr_pokemon = pokemon.corr(numeric_only=True)

# Ahora ya puedes ver la matriz sin errores
print(corr_pokemon)
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(corr_pokemon, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Mapa de Calor de Correlaciones en Pokémon')

plt.show()
```

### ¿la correlación entre qué valores se encuentra?
### la autocorrelación es la correlación de una variable consigo misma, ¿qué aprecias?
### Encuentra dos variables que veas que están fuertemente relacionadas según la correlación

Sp. Atk y Sp. Def: Suelen tener una relación notable (alrededor de 0.5). Los Pokémon que son fuertes en ataques especiales suelen tener también buena defensa especial. Parecido aplica a Attack y Defense.
La autocorrelacion siempre es 1.

# 27. Muestra una función distribución de probabilidad con kdeplot (curvas de nivel) de la defensa respecto al ataque.
### Di un bin 2D donde caigan muchos puntos (ataque, defensa).

```python
# Configuramos el estilo
sns.set_style('whitegrid')

# Creamos el mapa de densidades (curvas de nivel)
sns.kdeplot(x='Attack', y='Defense', data=pokemon, fill=True, cmap='Blues')

# Añadimos etiquetas y título
plt.title('Densidad de Probabilidad: Ataque vs Defensa')
plt.xlabel('Ataque')
plt.ylabel('Defensa')

plt.show()
```

# 28. Pinta una distribución de probabilidad conjunta con jointplot, pintando la defensa respecto al ataque del DataFrame original

```python
# Configuramos el estilo
sns.set_style('whitegrid')

# Creamos el jointplot
# Mostramos la relación entre Attack (eje x) y Defense (eje y)
sns.jointplot(x='Attack', y='Defense', data=pokemon, kind='scatter', color='purple')

# Añadimos un título (usando fig.suptitle porque jointplot crea su propia figura)
plt.gcf().suptitle('Distribución Conjunta: Ataque vs Defensa', y=1.02)

plt.show()
```
