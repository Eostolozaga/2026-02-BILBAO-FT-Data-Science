## Ejercicio 10
El amargor de las cervezas se mide por su nivel de IBUs.
1. ¿Cuál es el amargor medio de las cervezas?
2. ¿El máximo? ¿Y el mínimo?
3. ¿Qué cerveza es la que tiene el amargor máximo? ¿Cuál es la de amargor mínimo?

```python
amargor_medio = beers['ibu'].mean()
print(amargor_medio)
```

```python
amargor_max = beers['ibu'].max()
amargor_min = beers['ibu'].min()
print(f"Máximo: {amargor_max}, Mínimo: {amargor_min}")
```

```python
cerveza_max = beers.loc[beers['ibu'].idxmax(), 'name'] # Si hay varias con el mismo valor maximo tenemos que usar el .max
cerveza_min = beers.loc[beers['ibu'].idxmin(), 'name']

print(f"La más amarga es: {cerveza_max}")
print(f"La menos amarga es: {cerveza_min}")
```

## Ejercicio 11
No queda muy claro cuánto de dispar, de dispersa, es la medida de IBU. Mediante estadística descriptiva (una o dos métricas, sin gráficas), razona la dispersión de la variable IBU.

```python
std_ibu = beers['ibu'].std()

stats = beers['ibu'].describe()
iqr_ibu = stats['75%'] - stats['25%']

print(f"Desviación Estándar: {std_ibu}")
print(f"IQR: {iqr_ibu}")
```

## Ejercicio 20

## Ejercicio 24
Teniendo en cuenta la media de amargor de todas las cervezas, obtén una tabla con las cervezas cuyos estados tengan una media de amargor en cerveza por encima de la media

```python
media_global_ibu = birra_juntos['ibu'].mean()
media_global_ibu
```

```python
media_por_estado = birra_juntos.groupby('state')['ibu'].transform('mean')
media_por_estado
```

```python
birra_resultado = birra_juntos[media_por_estado > media_global_ibu]
birra_resultado
```
