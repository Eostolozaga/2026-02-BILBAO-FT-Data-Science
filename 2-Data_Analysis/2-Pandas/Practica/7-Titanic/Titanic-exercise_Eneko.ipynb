24. Using [`.pivot_table()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html) to count how many women or men survived by class, call it `counted`.

Don't panic and read the documentation!




```python
counted = titanic.pivot_table(index='sex',columns='pclass', values='survived', aggfunc='sum', margins = True)

counted
```

25. Add a new column with the sum of survived men and women, call it `counted['total']` 


```python
counted["total"] = counted.sum(axis=1)
counted
```

26. Sort `counted` by the `'total'` column. In which class the people survived the most?

```python
counted_sorted = counted.sort_values(by='total', ascending=False)

print(counted_sorted)
```

```python
mejor_clase = counted_sorted.index[0]
mejor_clase
```

31. Using `matplotlib` find the appropriate visualization to show the distribution of the column `'age'`

```python
import matplotlib.pyplot as plt
```

```python
# Configuramos el estilo y el tamaño
plt.figure(figsize=(10, 6))

# Creamos el histograma
# 'bins=30' divide el rango de edad en 30 barras
# 'edgecolor' ayuda a separar visualmente las barras
plt.hist(titanic['age'].dropna(), bins=30, color='skyblue', edgecolor='black')

# Añadimos títulos y etiquetas
plt.title('Distribución de edades en el Titanic', fontsize=15)
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Número de pasajeros', fontsize=12)

# Mostramos el gráfico
plt.show()
```

32. Using `matplotlib` find the appropriate plot to visualize the column `'gender'`

```python
gender_counts = titanic['sex'].value_counts()

plt.figure(figsize=(7, 5))
gender_counts.plot(kind='bar', color=['skyblue', 'salmon'], edgecolor='black')

plt.title('Distribución de Género en el Titanic', fontsize=14)
plt.xlabel('Género', fontsize=12)
plt.ylabel('Cantidad de Pasajeros', fontsize=12)
plt.xticks(rotation=0)

plt.show()
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
fallecidos_edad = titanic[titanic['survived'] == 0]['age'].dropna()

# 2. Creamos el gráfico de densidad (KDE)
plt.figure(figsize=(10, 6))
sns.kdeplot(fallecidos_edad, color="red", label="Not Survived", fill=True)

# 3. Personalización final
plt.title('Distribución de Edad de los No Supervivientes', fontsize=14)
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.legend()

plt.show()
```
